```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.model_selection import train_test_split

class PARAMS:
    Nclasses = 2
    List_Files = [r"C:\Users\DELL\Documents\PatternRecognition\assignment_01\non_linearly_seperable data\Class1.txt",r"C:\Users\DELL\Documents\PatternRecognition\assignment_01\non_linearly_seperable data\Class2.txt"]
    List_Class = [1, 2]


def main():
    # Reading Data
    list_X = []
    list_Y = []
    list_C = []
    for i in range(PARAMS.Nclasses):
        dat = readData(PARAMS.List_Files[i], PARAMS.List_Class[i])
        list_X = list_X + dat[0]
        list_Y = list_Y + dat[1]
        list_C = list_C + dat[2]
    
    Data = pd.DataFrame()
    Data['X'] = list_X
    Data['Y'] = list_Y
    Target = list_C

    # # Understanding the Data
    # print(Data.head())
    # Visualize(Data, Target)

    # Splitting the Data
    X_train, X_test, Y_train, Y_test = train_test_split(Data, Target, test_size=0.3, random_state=42)

    Model = BuildGausian()
    Model.fit(X_train, Y_train)
    Y_pred = Model.predict(X_test)
    Visualize(X_train,Y_train)
    Visualize(X_test, Y_pred)
    
    print(Incorrect(Y_test, Y_pred))


def Incorrect(test, pred):
    c = 0
    for i in range(len(test)):
        if (test[i] != pred[i]):
            c += 1
    
    return c

def Visualize(Data, Class):
    plt.scatter(Data['X'], Data['Y'], c=Class)
    plt.show()


def readData (filename, Class):
    allLines = open(filename, 'r').readlines()
    Data_X = []
    Data_Y = []
    Data_class = []
    for line in allLines:
        Data_X.append(list(map(float, line.strip().split()))[0])
        Data_Y.append(list(map(float, line.strip().split()))[1])
        Data_class.append(Class)
    
    return [Data_X, Data_Y, Data_class]


class BuildGausian():
    def fit(self, X, Y):
        self.Class, self.ClassCounts = np.unique(Y, return_counts=True)
        self.NClasses = len(self.Class)

        self.prior = {self.Class[i]: self.ClassCounts[i] for i in range(len(self.Class))}
        print(self.prior)
        ClassBasedData = {}
        for Class in self.Class:
            ClassBasedData[Class] = [[],[]]
        
        for i in range(len(Y)):
            ClassBasedData[Y[i]][0].append(list(X.iloc[i])[0])
            ClassBasedData[Y[i]][1].append(list(X.iloc[i])[1])
        print(ClassBasedData[1][0][:5]) 

        self.pdfs = {}
        for Class in self.Class:
            Data = np.array(ClassBasedData[Class])
           
            Mean = np.mean(Data, axis=1)
            Cov = np.cov(Data)

            self.pdfs[Class] = {}
            self.pdfs[Class]['Mean'] = Mean
            self.pdfs[Class]['Covariance'] = Cov
        print(self.pdfs[1]["Mean"])
    def predict(self, X):
        Y = []
        for row in X.values:
            Proba = {}
            for C in self.Class:
                Proba[C] = self.Posterior(row, C) * self.prior[C]
            
            res = self.Class[0]
            for C in self.Class:
                if (Proba[res] < Proba[C]):
                    res = C

            Y.append(res)
        return Y
    
    def Posterior(self, Data, Class):
        pdf = self.pdfs[Class]
        Mean = pdf['Mean']
        Cov = pdf['Covariance']
        Data = np.array(Data)
        
        Num = np.exp(-0.5 * np.matmul(np.matmul(np.transpose(Data - Mean), np.linalg.inv(Cov)), (Data - Mean)))
        Den = (2*3.14) * ((np.linalg.det(Cov)) ** 0.5)      
        Ans = Num/Den
        return Ans
   

main()
```
